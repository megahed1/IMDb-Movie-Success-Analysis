```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv("Movies Dataset.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df['revenue'] = df['revenue'].astype('float')
```

```python
df['release_date'] =pd.to_datetime(df['release_date'])
```

```python
df.describe()
```

## Cleaning

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.isna().sum()
```

```python
df.drop(columns = 'homepage', inplace = True) 
```

```python
df.columns
```

```python
df.drop(columns = ['tagline', 'keywords'], inplace = True)
```

```python
df['production_companies'] = df['production_companies'].fillna("Unkown company")
```

```python
remove_row_nulls = ['cast', 'director', 'overview', 'genres']
```

```python
for row in remove_row_nulls:
    df = df.dropna(subset=row)
```

```python
sns.boxplot(df['revenue_adj'])
plt.grid(True)
plt.title('Outliers in Revenue Adjustment')
plt.show()
```

```python
top_3_movies = df[df['revenue_adj'] > 3500000000]
top_3_movies
```

## Analysis

```python
corr_matrix = df[['budget_adj', 'revenue_adj', 'vote_average']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
```

```python
df_1997 = df[df['release_year'] >= 1997]
```

```python
df_1997
```

```python
top_director = (
    df_1997.groupby('director')['revenue_adj']
        .sum()
        .reset_index()
        .sort_values(by = 'revenue_adj', ascending = False)
        .head(10)
)
top_director
```

```python
sns.barplot(x = 'director', y = 'revenue_adj', data=top_director, palette= 'cubehelix')
plt.title("top 10 directors by total revenue adjustment")
plt.xlabel("Director names")
plt.ylabel("Revenue")
plt.gca().tick_params(axis='x', rotation=-45)
plt.show()
```

```python
df["rounded_vote"] = df['vote_average'].round()
```

```python
sns.countplot(x="rounded_vote", data = df)
plt.title("how do audiences rate Movies?")
```
